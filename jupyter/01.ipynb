# L1: The Geometry of Linear Equations
$Ax$: Combinations of Columns of A

e.g 1<br>
$
\left\{
\begin{align*}
2x - y &= 0 \\
-x + 2y &= 3 \\
\end{align*}
\right.
$

We can view the equations through **row picture** and **column picture**

Row picture is by thinking of the equation being the intersection of two lines: $2x-y=0$ and $-x+2y=3$
<br> Below is the visualization:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x-values range
x = np.linspace(-5, 5, 100)

# Equation 1: 2x - y = 0
y1 = 2*x

# Equation 2: -x + 2y = 3
y2 = (3 + x) / 2

# Plot the lines
plt.plot(x, y1, label='2x - y = 0')
plt.plot(x, y2, label='-x + 2y = 3')

# Plot x-axis (y = 0)
plt.axhline(0, color='black', linewidth=0.5)

# Plot y-axis (x = 0)
plt.axvline(0, color='black', linewidth=0.5)

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs of the Equations')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

column picture: <br>
$x\left[\begin{matrix}2 \\-1 \\\end{matrix}\right]
+
y\left[\begin{matrix}-1 \\2 \\\end{matrix}\right]
=
\left[\begin{matrix} 0 \\3 \\\end{matrix}\right]$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
v1 = np.array([2, -1])
v2 = np.array([-1, 2])

# Create the figure and axis
fig, ax = plt.subplots()

# Plot the x-axis
ax.axhline(0, color='black', linewidth=0.5)

# Plot the y-axis
ax.axvline(0, color='black', linewidth=0.5)

# scale coefficient of v1 and v2
scale_v1, scale_v2 = 2, 1.5

# Extend the vectors by scaling
scaled_v1 = v1 * scale_v1
scaled_v2 = v2 * scale_v2

# Plot the vectors
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector v1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector v2')

# Plot the scaled vectors
ax.quiver(0, 0, scaled_v1[0], scaled_v1[1], angles='xy', scale_units='xy', scale=1, color='blue', linestyle='dotted')
ax.quiver(0, 0, scaled_v2[0], scaled_v2[1], angles='xy', scale_units='xy', scale=1, color='red', linestyle='dotted')

# Calculate the result of vector addition
result = scaled_v1 + scaled_v2

# Plot the result vector
ax.quiver(0, 0, result[0], result[1], angles='xy', scale_units='xy', scale=1, color='green', label=f'Result ({scale_v1}v1 + {scale_v2}v2)')

# Plot the parallelogram using scaled_v1 and scaled_v2
parallelogram = np.column_stack((scaled_v1, scaled_v2, scaled_v1 + scaled_v2))
ax.fill(parallelogram[0], parallelogram[1], alpha=0.3, color='gray')

# Set the x and y limits
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Visualization of Vectors')

# Add legend
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```

e.g 2<br>
$$
\left\{
\begin{aligned}
2x -& y & = 0 \\
-x +& 2y - z & = -1 \\
-& 3y + 4z & = 4\\
\end{aligned}
\right.
$$
<br>
below is a row picture:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.linalg import solve

# Create a grid of points
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

# Define the three functions (representing three planes)
z1 = 2 * x
z2 = -x + 2*y - 1
z3 = (3*y + 4) / 4

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the functions
ax.plot_surface(x, y, z1, color='r', alpha=0.5, rstride=100, cstride=100)
ax.plot_surface(x, y, z2, color='b', alpha=0.5, rstride=100, cstride=100)
ax.plot_surface(x, y, z3, color='g', alpha=0.5, rstride=100, cstride=100)

# Define the system of equations in matrix form to find intersection points
A = np.array([[2, -1, 0], [-1, 2, -1], [0, -3, 4]])
b = np.array([0, -1, 4])

# Solve the system of equations
X = solve(A, b)

# Add intersection point to the plot
ax.scatter(X[0], X[1], X[2], color='black')
ax.text(X[0], X[1], X[2], "Intersection point", color='black')

# Give labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Set title
ax.set_title("3D plot of the system of equations")

# Show the plot
plt.show()
```

column picture: <br>
$x\left[\begin{matrix}2 \\-1 \\0 \\\end{matrix}\right]
+
y\left[\begin{matrix}-1 \\2 \\-3 \\\end{matrix}\right]
+
z\left[\begin{matrix} 0 \\-1 \\4 \\\end{matrix}\right]
=\left[\begin{matrix} 0 \\-1 \\4 \\\end{matrix}\right]$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Original vectors
v1 = np.array([2, -1, 0])
v2 = np.array([-1, 2, -3])
v3 = np.array([0, -1, 4])

# Random extensions
np.random.seed(0)  # To make the random factors reproducible
factors = np.random.rand(3)
v1_ext = factors[0] * v1
v2_ext = factors[1] * v2
v3_ext = factors[2] * v3

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the original vectors
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='b', alpha = 0.5)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='r', alpha = 0.5)
ax.quiver(0, 0, 0, v3[0], v3[1], v3[2], color='g', alpha = 0.5)

# Plot the extended vectors
ax.quiver(0, 0, 0, v1_ext[0], v1_ext[1], v1_ext[2], color='b')
ax.quiver(0, 0, 0, v2_ext[0], v2_ext[1], v2_ext[2], color='r')
ax.quiver(0, 0, 0, v3_ext[0], v3_ext[1], v3_ext[2], color='g')

# Set the limits
ax.set_xlim([min(0, v1_ext[0], v2_ext[0], v3_ext[0]), max(0, v1_ext[0], v2_ext[0], v3_ext[0])])
ax.set_ylim([min(0, v1_ext[1], v2_ext[1], v3_ext[1]), max(0, v1_ext[1], v2_ext[1], v3_ext[1])])
ax.set_zlim([min(0, v1_ext[2], v2_ext[2], v3_ext[2]), max(0, v1_ext[2], v2_ext[2], v3_ext[2])])

# Adjust the viewing angle
ax.view_init(elev=20, azim=-35)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```
